# Probability simulation

This is exercise 5.2 from _Regression and Other Stories_.

```r
# Params
pr_male <- 0.48
man_log_weight_mean <- 5.13
man_log_weight_std <- 0.17
woman_log_weight_mean <- 4.96
woman_log_weight_std <- 0.20
num_adults <- 10
weight_limit <- 1750
```

```r
num_sims <- 1000
total_weight <- rep(NA, num_sims)
for (s in 1:num_sims) {
    male <- rbinom(num_adults, 1, pr_male)
    log_weight <- ifelse(
        male == 1,
        # Generate num_adults log weights for men
        rnorm(num_adults, man_log_weight_mean, man_log_weight_std),
        # Generate num_adults log weights for women
        rnorm(num_adults, woman_log_weight_mean, woman_log_weight_std))
    weight <- exp(log_weight)
    total_weight[s] <- sum(weight)
}
```
