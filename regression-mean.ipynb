# Regression to the mean 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(42)
```

```python
n = 1000
skill = rng.normal(50, 10, n)
luck_midterm = rng.normal(0, 10, n)
luck_final = rng.normal(0, 10, n)
score_midterm = skill + luck_midterm
score_final = skill + luck_final
```

```python
m, b = np.polyfit(score_midterm, score_final, 1)
m, b
```

```python
plt.figure(figsize=(8, 8))
plt.title("Test scores")
plt.xlabel("Midterm")
plt.ylabel("Final")
plt.scatter(x=score_midterm, y=score_final, s=0.5)
plt.plot(score_midterm, m * score_midterm + b, color="red")
plt.show()
```
